## Task 11: Scatter Plot

Create a scatter plot that shows the correlation between total interactions and likes.

```python
# call_on_students(1)
```

```python
# Your code here
fig, ax = plt.subplots(figsize=(16, 8))

ax.scatter(x=df['Total Interactions'], y=df['like'])
ax.set_xlabel("Total Interactions")
ax.set_ylabel("Likes")
```

Let's discuss: what do you notice?

- It appears that there is a very strong positive linear relationship between total interactions and number of likes. This makes sense because number of likes is a direct input in the total interactions equation, as like is a type of interaction. We can also see there is one main clear outlier in the datset that recieved far more interactions and likes in comparison to the rest of the dataset.

## Task 13: Loop to List

```python
# call_on_students(1)
```

```python
# Your code here
first_twenty_songs[0].keys()

track_names = []
for song in first_twenty_songs:
    track_names.append(song['name'])

track_names
```

## Task 14: Create a new dictionary

Create a dictionary called `song_dictionary` which consists of each track name `string` as a key and a `tuple` of artists associated with each track as a value.

```python
# call_on_students(1)
song_dictionary = {}
artist_list_2 = []
for song in first_twenty_songs:
    artist_list = []
    for x in song['artists']:
        artist_list.append(x['name'])
    artist_list_2.append(tuple(artist_list))
    
artist_list_2
```
